# 02 - Exploratory Data Analysis

Sanity-check the simulator output: ripening curves, gas accumulation, temperature effect on shelf life.

```python
import sys, pathlib
ROOT = pathlib.Path.cwd().resolve().parents[1]
sys.path.insert(0, str(ROOT))
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv(ROOT / 'data' / 'processed' / 'banana_synthetic_v1.csv')
df.shape
```

```python
fig, ax = plt.subplots(2, 2, figsize=(12,8))
for regime in df['regime'].unique():
    g = df[df['regime']==regime].groupby('hours_since_harvest')[['co2_ppm','ethylene_ppm','days_until_spoilage','ripeness_estimate']].mean().reset_index()
    ax[0,0].plot(g['hours_since_harvest'], g['co2_ppm'], label=regime)
    ax[0,1].plot(g['hours_since_harvest'], g['ethylene_ppm'], label=regime)
    ax[1,0].plot(g['hours_since_harvest'], g['days_until_spoilage'], label=regime)
    ax[1,1].plot(g['hours_since_harvest'], g['ripeness_estimate'], label=regime)
ax[0,0].set_title('CO2 ppm vs time'); ax[0,0].legend(fontsize=8)
ax[0,1].set_title('Ethylene ppm vs time'); ax[0,1].legend(fontsize=8)
ax[1,0].set_title('RSL (days) vs time')
ax[1,1].set_title('Ripeness vs time')
plt.tight_layout(); plt.show()
```
